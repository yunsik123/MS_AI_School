### 구조방정식: 측정모델(확인적요인분석)+구조모델(다중회귀분석)

```r
#외생잠재변수와 내생잠재변수가 있으며 
#1단계: 측정모델 : 단일차원성분석 신뢰도분석 타당도 분석
#2단계: 구조모델 : 경로분석
```

```r
#단일차원성분석-적합도
#1.절대적합도
#2.증분적합도
#3.간명적합도
#이중 한개를 골라서 해야하며
#단일 차원성을 충족하기 위해서는 일반적으로 0.6이상의 표준화 요인적재값을 가져야함
#적합도 지표가 적정수준에 도달할때까지 씀

#모델 적합도 수준이 부족하다면 적합도 개선을 위해 수정지표를 쓰는데
#새로운 모수를 도입하거나 불필요한 모수를 제거함.
```

```r
#신뢰도는 측정척도가 측정하려고 의도하는 것을 얼마나 정확하게 오차 없이 측정하고있는지 나타냄
#크론바흐 알파계수 0.7보다 커야함
#복합신뢰도 0.7보다 커야함
#평균분산추출 0.5보다커야함

#이걸 만족하면 잠재변수가 좋음
```

```r
#타당도
#타당도는 측정척도가 측정하려고 의도하는 것을 얼마나 충실하게 측정하고있는가
#1. 집중타당도(요인적재값이 클수록 크다, 평균분산추출은 0.5이상이여야한다.)
#2. 판별타당도: 다른 잠재변수와는 관련을 갖지 않는정도이다.
```

```r
#구조모델: 경로계수의 부호는 가설화된 관계의 방향과 일치한지 결정계수는 적정한지 평가
```

```r
library(lavaan)
str(PoliticalDemocracy)#11개의 관측변수는 3개의 잠재변수를 측정하도록 고려됨
#11개는 진짜로 측정한거며
#3개는 11개를 통해 정의될수있는것임.
#?PoliticalDemocracy
#경로도로 표현할수있으며 

cfa<-  "ind60 =~x1+x2+x3
        dem60 =~y1+y2+y3+y4
        dem66 =~y5+y6+y7+y8 "  #각행은 하나의 잠재변수를 나타

cfa(model=cfa, data=PoliticalDemocracy)
fit<-cfa(model=cfa, data=PoliticalDemocracy)
summary(fit,fit.measures=TRUE, standardized=TRUE)# 3개의 큰 섹션값이 나옴

# 첫번째 헤더 색션은 모델모수개수 관측값개수 추정방법 통계량 p값 적합도검정결과

# 두번째는 추가적인 것으로 모수처정치 공분산 잔차분산 등 값

parameterEstimates(fit,standardized=TRUE)#추정치 파라미터  ~~는 공분산관계 =~는 잠재변수관계
standardizedsolution(fit)#표준화된 추정치임

coef(fit)#잘못측정된거만 보여줌

#확인적요인분석의 목표는 관측변수간의 관계를 보려고한거
```

```r
residuals(fit,type="cor")$cov #상관계수 잔차행렬임




#적합도지표
fitMeasures(fit, c("chisq","df","pvalue","gfi","rmsea","cfi"))


lavInspect(fit,what="rsquare")

#구조모델 경로도
library(semPlot)
semPaths(fit,what="std", layout="tree2",
        edge.label.cex=1, edge.color='royalblue',
        color=list(lat="lightcoral", man="lavenderblush"),
        fade=FALSE, style="lisrel",curvature=2)

#신뢰도
library(semTools)
reliability(fi2)


#매개효과분석
#매개효과는 총효과=직접효과+간접효과
```
