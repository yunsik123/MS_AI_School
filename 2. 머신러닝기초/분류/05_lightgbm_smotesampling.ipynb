```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split

train,test= train_test_split(creditcard,test_size=0.2, random_state=101)
```

```python
test['Class'].value_counts(normalize=True)
```

```python
X_train = train.drop(['Class'], axis=1)
y_train = train['Class']
X_test = test.drop(['Class'], axis=1)
y_test = test['Class']
```

```python
#pip install imbalanced-learn
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import confusion_matrix
 
model = LGBMClassifier(
  tree_learner = 'serial',
  # 운영 파라미터
  n_jobs = -1, force_col_wise= True, verbose=0,
  # 모델 파라미터
  n_estimators = 100,
  num_leaves = 255, learning_rate = 0.1,
  subsample = 0.7,
  random_state=42
  )
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
pd.crosstab(y_test,y_pred)
#정확도는 좋지만 불균형데이터라 좋은거임(과적합?)
#밑에 table을 보면 알수있음
```

```python
#smotesampling써보자.
#비슷한애들을 만드는거임 
#knn쓰면 바로옆에있는 애들따라서만드는것
#minority는 소수클래스 쓰겠다는거임
from imblearn.over_sampling import SMOTE

smote_sampler=SMOTE(sampling_strategy='minority',k_neighbors=1)
X_train_smote, y_train_smote = smote_sampler.fit_resample(X_train,y_train)
```

```python
X_train_smote.shape,y_train_smote.shape
```

```python
sns.countplot(x=y_train_smote)
```

```python
y_train_smote.value_counts(normalize=True)#normalize를 이용한 상대빈도
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import confusion_matrix
 
model1 = LGBMClassifier(
  tree_learner = 'serial',
  # 운영 파라미터
  n_jobs = -1, force_col_wise= True, verbose=0,
  # 모델 파라미터
  n_estimators = 100,
  num_leaves = 255, learning_rate = 0.1,
  subsample = 0.7,
  random_state=42
  )
```

```python
model1.fit(X_train_smote, y_train_smote)
```

```python
y_pred_smote=model1.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_smote)
```

```python
pd.crosstab(y_test, y_pred_smote)
```

```python
from sklearn.metrics import precision_score #정밀도였나?
precision_score(y_test,y_pred_smote)#오버샘플링후 정밀도
```
