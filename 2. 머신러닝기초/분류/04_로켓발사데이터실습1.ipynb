```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
launch_data=pd.read_csv('RocketLaunchDataCSV.csv')
launch_data.head()
```

```python
launch_data.info()
```

```python
launch_data.isnull().sum() 
#아래열은 너무 많은 결측치가 있음으로 없애준다.
temp=['Name','Time (East Coast)','Crewed or Uncrewed','Crewed or Uncrewed','Temp at Launch Time','Date',
          'Wind Speed at Launch Time','Hist Ave Max Wind Speed','Hist Ave Visibility','Hist Ave Sea Level Pressure','Notes']
```

```python
launch_data=launch_data.drop(columns=temp,axis=1)
```

```python
launch_data.info()
```

```python
#결측치값채워주기
launch_data['Launched?']=launch_data['Launched?'].fillna('N')
launch_data.info()
```

```python
launch_data=launch_data.dropna()
launch_data.info()
```

```python
launch_data.describe()
```

```python
# split data 
X = launch_data.drop(columns=['Launched?'],axis=1)
y = launch_data['Launched?']
```

```python
#X값중 object값을 라벨인코딩해줘야한다.
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()
X['Location']=label_enc.fit_transform(X['Location'])
X['Wind Direction']=label_enc.fit_transform(X['Wind Direction'])
X['Sea Level Pressure']=label_enc.fit_transform(X['Sea Level Pressure'])
X['Day Length']=label_enc.fit_transform(X['Day Length'])
X['Condition']=label_enc.fit_transform(X['Condition'])
```

```python
#train-test나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=101)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
#스케일링
scaler=StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV

cross_acc = cross_val_score(rf, X_train, y_train, cv=3)
cross_acc
```

```python
param_grid = {'max_depth': [3,5,7], 
            'min_samples_split': [2,3,4]} 

gr_search = GridSearchCV(RandomForestClassifier(random_state=101), param_grid, cv=3) 
gr_search 
```

```python
gr_search.fit(X, y) 
gr_search.best_params_ 
```
