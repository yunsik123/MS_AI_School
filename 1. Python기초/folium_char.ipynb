```python
import folium

map = folium.Map(location=[37.4965,126.9573],
                zoom_start=17,
                width=750,
                height=500)

folium.Marker(location=[37.4965,126.9573],
             popup='숭실대학교',
             tooltip='숭실대학교 정문',
             icon=folium.Icon(color='red',icon='horse',prefix='fa')).add_to(map)

#map.save('map.html')
```

```python
import pandas as pd

df=pd.read_excel("반려견_놀이터.xlsx")
df.columns
```

```python
#mean값을 중심으로 그려진다고 생각하삼

dogparkmap = folium.Map(location=[df['Latitude'].mean(),df['Longitude'].mean()],
                zoom_start=17,
                width=750,
                height=500)
for n, row in df.iterrows():
    folium.Marker(location=[row['Latitude'],row['Longitude']],
                 popup='숭실대학교',
                 icon=folium.Icon(color='red',icon='horse',prefix='fa')).add_to(dogparkmap)
dogparkmap
#dogparkmap.save('dogparkmap.html')
```
