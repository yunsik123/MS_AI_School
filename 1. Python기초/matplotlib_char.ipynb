```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("fine_dust.xlsx")
```

```python
data2018=data[2018]
plt.figure(figsize=(15,4))
plt.plot(data2018,color='b',marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 fine duset line graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,4))

for year in range(2016,2020):
    chartdata=data[year]
    plt.plot(chartdata,marker='o',label=year)
```

```python
data2017 = data[2017]
plt.plot(data2017,color='g')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2017 fine duset line graph')
plt.grid()
plt.show()
```

```python
data2017 = data[2017]
plt.bar(data2017.index,data2017,color='g')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2017 fine duset line graph')
plt.grid()
plt.show()
```

```python
data=pd.read_excel("health_screenings_2020_1000ea.xlsx")
```

```python
data.info()
```

```python
heightdata=data['height']
```

```python
systolic_data=data['systolic']
diastolic_data=data['diastolic']
blood_sugar_data=data['blood_sugar']

plt.figure(figsize=(10,6))

plt.scatter(systolic_data,diastolic_data,color='r',label='systolic*diastolic')
plt.scatter(systolic_data,blood_sugar_data,color='g',label='systolic*blood_sugar_data')
plt.scatter(diastolic_data,blood_sugar_data,color='b',label='diastolic*blood_sugar_data')

plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 Health Screenings Group Scatter Graph')
plt.legend()
plt.show()
```

```python
female_data=data.loc[data['gender']==2,['gender','height']]
male_data=data.loc[data['gender']==1,['gender','height']]

plt.figure(figsize=(10,6))
plt.hist(female_data['height'],bins=6,label='Female')
plt.hist(male_data['height'],bins=6,label='male')


plt.xlim(130,180)
plt.title('title')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

height = np.array([male_data['height'],female_data['height']],dtype='object')

plt.figure(figsize=(10,6))
plt.boxplot(height,labels=['Male','Female'])

plt.ylim(130,200)
plt.title('title')
plt.show()
```
